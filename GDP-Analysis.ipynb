```python
# Creating Dictionary for States with State abbreviation as key and state name as value
dict_state = {'AP' : ['Andhra_Pradesh'],
              'AR' : ['Arunachal_Pradesh'],
              'AS' : ['Assam'],
              'BR' : ['Bihar'],
              'CG' : ['Chhattisgarh'],
              'GA' : ['Goa'],
              'GJ' : ['Gujarat'],
              'HR' : ['Haryana'],
              'HP' : ['Himachal_Pradesh'],
              'JH' : ['Jharkhand'],
              'KA' : ['Karnataka'],
              'KL' : ['Kerala'],
              'MP' : ['Madhya_Pradesh'],
              'MH' : ['Maharashtra'],
              'MN' : ['Manipur'],
              'ML' : ['Meghalaya'],
              'MZ' : ['Mizoram'],
              'NL' : ['Nagaland'],
              'OR' : ['Odisha'],
              'PB' : ['Punjab'],
              'RJ' : ['Rajasthan'],
              'SK' : ['Sikkim'],
              'TN' : ['Tamil_Nadu'],
              'TS' : ['Telangana'],
              'TR' : ['Tripura'],
              'UP' : ['Uttar_Pradesh'],
              'UK' : ['Uttarakhand']}
```

```python
# Importing library to read filenames using wild characters
import glob

# Traversing dictionary and adding dataframe for each state by reading CSV
# Added error handling to resolve file encoding issue
for state in dict_state.items():
    filepath = glob.glob("./NAD-" + state[1][0] + "*.csv")
    try:
        # Reading csv file
        df_state = pd.read_csv(filepath[0])
    except:
        # exception handling to resolve encoding issue
        df_state = pd.read_csv(filepath[0],encoding = "latin-1")
    # Updating dictionary with data read from CSV
    state[1].append(df_state)
```

```python
# Sample of dictionary whether dataframe is updated or not
print(dict_state['AP'][1].shape)
print(dict_state['MH'][1].head(5))
print(dict_state['MN'])
```

```python
# Cleansing and combinig dataframes, Operations performed are as below
# Cleansing datafram - remove sub-groups in sub-sectors,remove all columns except S.No., Item and 2014-15
# Update column name with State name
# Merging Dataframes

# Consolidated Dataframe for state GDP
df_GDP_State = pd.DataFrame()

for state in dict_state.items():
    # Cleansing rows
    state[1][1].drop(state[1][1][state[1][1]['S.No.'].str.contains('\.\d')].index, inplace = True)
    
    # Cleansing columns
    state[1][1] = state[1][1][['Item','2014-15']]
    
    # Updating index and renaming column to state name
    state[1][1].rename(columns={'2014-15' : state[1][0]}, inplace = True)

    # Merge dataframes
    if df_GDP_State.empty:
        df_GDP_State = state[1][1]
    else:
        df_GDP_State = pd.merge(df_GDP_State, state[1][1], how='inner', on='Item')
    
    # Printing State name and shape of dataframe of state
    print(state[1][0])
    print(state[1][1].shape)
    
# End of loop
```

```python
# Analysing merged dataframe
print(df_GDP_State.shape)
print(df_GDP_State.info())
print(df_GDP_State)
```

```python
# Dataframe for GDP per capita for all states
df_GDP_per_Capita = df_GDP_State[df_GDP_State['Item'] == 'Per Capita GSDP (Rs.)']

# Renaming Item column and updating it as Index
df_GDP_per_Capita.rename(columns={'Item' : 'State'}, inplace = True)
df_GDP_per_Capita.set_index('State', inplace = True)

# Transposing dataframe to plot
df_GDP_per_Capita = df_GDP_per_Capita.transpose()

# Sorting dataframe
df_GDP_per_Capita.sort_values(by='Per Capita GSDP (Rs.)', ascending = False, inplace = True)
```

```python
# Performing necessary transformations to add category data
df_GDP_State.rename(columns={'Item' : 'State'}, inplace = True)
df_GDP_State.set_index('State', inplace = True)
df_GDP_State = df_GDP_State.transpose()
```

```python
# Adding column to store Category Values
# Allocating last quartile to all records
df_GDP_State['Category'] = 'C4'

# Update category C4 - (0-0.2), C3 - (0.2-0.5), C2 - (0.5-0.85), C1 - (0.85-1)
df_GDP_State.loc[df_GDP_State['Per Capita GSDP (Rs.)'] >  df_GDP_State['Per Capita GSDP (Rs.)'].quantile(0.2),'Category'] = 'C3'
df_GDP_State.loc[df_GDP_State['Per Capita GSDP (Rs.)'] >  df_GDP_State['Per Capita GSDP (Rs.)'].quantile(0.5),'Category'] = 'C2'
df_GDP_State.loc[df_GDP_State['Per Capita GSDP (Rs.)'] >  df_GDP_State['Per Capita GSDP (Rs.)'].quantile(0.85),'Category'] = 'C1'
```

```python
# Check Category update
df_GDP_State.sort_values(by = 'Category')
```

```python
# Creating dictionary of Sector and sub-sector
# Creating list of sub-sector
dict_Sector = {'Primary' : [], 'Secondary' : [], 'Tertiary' : []}
# Creating list of Sub-Sector as well
list_sub_sector = []
sector_key = ''
for fields in list(df_GDP_State)[::-1]:
    if fields in dict_Sector.keys():
        sector_key = fields
    else:
        if sector_key:
            dict_Sector[sector_key].append(fields)
            list_sub_sector.append(fields)
            
# Appending 'Gross State Domestic Product' to Sub Category list
list_sub_sector.append('Gross State Domestic Product')
            
# Check dictionary category and sub sector mapping
print(dict_Sector)
print(list_sub_sector)
```

```python
# Dataframe grouped Category with aggregate function of sum
df_Category_Pivot = pd.pivot_table(df_GDP_State,index=["Category"],values = list_sub_sector,aggfunc=np.sum)
df_Category_Pivot
```

```python
# Creating dataframe with percentage sub category of GSDP 
df_Category_Pivot_perc = pd.DataFrame()
for field in list_sub_sector:
    if field != 'Gross State Domestic Product':
        df_Category_Pivot_perc[field] = (df_Category_Pivot[field]/df_Category_Pivot['Gross State Domestic Product']*100).round(2)

df_Category_Pivot_perc
```

```python
# Transposing dataset to perform further calculation
df_Category_Pivot_perc = df_Category_Pivot_perc.transpose()
df_Category_Pivot_perc
```

```python
# Adding Cumulative sum column to find sub-categories contributing to 80% of GSDP
# Extracting sub-sectors which contribute to 80% of GSDP for each category along with ranking
dict_cat_subsector = {'C1':{}, 'C2':{}, 'C3':{}, 'C4':{}}
for col in df_Category_Pivot_perc.columns.values:
    newCol = col + ' Cumulative Sum'
    df_Category_Pivot_perc.sort_values(by = col, ascending = False, inplace = True)
    df_Category_Pivot_perc[newCol] = df_Category_Pivot_perc[col].cumsum(axis=0)
    
    rank = 1
    for index, row in df_Category_Pivot_perc.iterrows():
        if row[newCol] < 80:
            dict_cat_subsector[col].update({rank : index})
        else:
            dict_cat_subsector[col].update({rank : index}) # Just add values that crosses 80
            break
        rank = rank +1

# printing dataframe
df_Category_Pivot_perc
```

```python
# Checking dictionary holding subsector contribution to total GSDP for each category
dict_cat_subsector
```

```python
# creating dataframe for top 3/4/5 sub-sectors which contribute to 80% GSDP for each category
df_top_subsec = pd.DataFrame()
df_top_subsec['Rank'] = [1,2,3,4,5]
for rec_cat, dict_sec in dict_cat_subsector.items():
    list_top5 = []
    for rec_sec in dict_sec.items():
        if rec_sec[0] <=5:
            list_top5.append(rec_sec[1])
    df_top_subsec[rec_cat] = list_top5
df_top_subsec.set_index('Rank', inplace = True) 

# Dataframe with top3/4/5 subsecotrs contributing to 80% of GSDP
df_top_subsec
```

```python
# Load dataset for 'Education drop-out rates'
df_Dropout_Rates = pd.read_csv("./rs_session243_au570_1.1.csv")
# check dataset loaded
print(df_Dropout_Rates.shape)
print(df_Dropout_Rates.info())
```

```python
df_Dropout_Rates
```

```python
# Cleansing State names, updating space in state name with space, to bring in line with our previous dataframe
df_Dropout_Rates.replace(' ','_',regex=True, inplace = True)

# Updating incorrect name
df_Dropout_Rates.loc[df_Dropout_Rates['Level of Education - State']=='Chhatisgarh','Level of Education - State'] = 'Chhattisgarh'
df_Dropout_Rates.loc[df_Dropout_Rates['Level of Education - State']=='Uttrakhand','Level of Education - State'] = 'Uttarakhand'


df_Dropout_Rates
```

```python
# Creating list of State name used in previous dataframes
list_statename = []
for dict_rec in dict_state.values():
    list_statename.append(dict_rec[0])
list_statename
```

```python
# Cleansing data and retaining only for states which were used in previous analysis
df_Dropout_Rates = df_Dropout_Rates[df_Dropout_Rates['Level of Education - State'].isin(list_statename)]
print(df_Dropout_Rates.shape)
df_Dropout_Rates
```

```python
# Updating index
df_Dropout_Rates.set_index('Level of Education - State', inplace=True)
```

```python
# Cleansing columns only retaining for 2014-15
list_col_names = [col for col in df_Dropout_Rates.columns if '2014-2015' in col]

# Removing duplicate entry to Primary 2014-2015
list_col_names.remove('Primary - 2014-2015.1')
list_col_names
```

```python
# Updating dataframe
df_Dropout_Rates = df_Dropout_Rates.loc[:,list_col_names]
print(df_Dropout_Rates.shape)
df_Dropout_Rates
```

```python
# Number of NULL values in each row
df_Dropout_Rates.isnull().sum(axis=1)

# none of the row has all NULL values
```

```python
# Add GSDP and Per Capita GSDP from previous dataframes to current dataframe
df_Dropout_Rates = pd.merge(left=df_Dropout_Rates, left_index=True, right=df_GDP_State.loc[:,['Per Capita GSDP (Rs.)','Category']], right_index=True, how='inner')
```

```python
# Check Dataframe
df_Dropout_Rates
```

```python
# Correlation matrix for all states
df_Corr = df_Dropout_Rates.loc[:,df_Dropout_Rates.columns!='Category'].corr()
df_Corr
```

```python
# Plotting correlation of education dropout rates with per capits GSDP
plt.figure()
plt.style.use('default')
sns.heatmap(df_Corr, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, linewidth=0.5, annot=True)
plt.title('Correlation of education dropout rates with per capita GSDP', fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.show()
```

### There is negative correlation between education dropout rates and per capita GSDP

```python
# Correlation matrix for each Category
df_Corr_C1 = df_Dropout_Rates.loc[df_Dropout_Rates['Category']=='C1',df_Dropout_Rates.columns!='Category'].corr()
df_Corr_C2 = df_Dropout_Rates.loc[df_Dropout_Rates['Category']=='C2',df_Dropout_Rates.columns!='Category'].corr()
df_Corr_C3 = df_Dropout_Rates.loc[df_Dropout_Rates['Category']=='C3',df_Dropout_Rates.columns!='Category'].corr()
df_Corr_C4 = df_Dropout_Rates.loc[df_Dropout_Rates['Category']=='C4',df_Dropout_Rates.columns!='Category'].corr()
```

```python
# Plotting correlation for each Category
plt.figure(figsize = (20,40))
plt.style.use('default')

plt.subplot(411)
hm1 = sns.heatmap(df_Corr_C1, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, linewidth=0.5, annot=True, annot_kws={"size": 18})
hm1.set_xticklabels(hm1.get_xticklabels(), rotation = 0, fontsize = 12)
hm1.set_yticklabels(hm1.get_yticklabels(), rotation = 0, fontsize = 12)
plt.title('Category C1 Correlation', fontdict={'fontsize': 20, 'fontweight': 'bold'})

plt.subplot(412)
hm2 = sns.heatmap(df_Corr_C2, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, linewidth=0.5, annot=True, annot_kws={"size": 18})
hm2.set_xticklabels(hm2.get_xticklabels(), rotation = 0, fontsize = 12)
hm2.set_yticklabels(hm2.get_yticklabels(), rotation = 0, fontsize = 12)
plt.title('Category C2 Correlation', fontdict={'fontsize': 20, 'fontweight': 'bold'})

plt.subplot(413)
hm3 = sns.heatmap(df_Corr_C3, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, linewidth=0.5, annot=True, annot_kws={"size": 18})
hm3.set_xticklabels(hm3.get_xticklabels(), rotation = 0, fontsize = 12)
hm3.set_yticklabels(hm3.get_yticklabels(), rotation = 0, fontsize = 12)
plt.title('Category C3 Correlation', fontdict={'fontsize': 20, 'fontweight': 'bold'})

plt.subplot(414)
hm4 = sns.heatmap(df_Corr_C4, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, linewidth=0.5, annot=True, annot_kws={"size": 18})
hm4.set_xticklabels(hm4.get_xticklabels(), rotation = 0, fontsize = 12)
hm4.set_yticklabels(hm4.get_yticklabels(), rotation = 0, fontsize = 12)
plt.title('Category C4 Correlation', fontdict={'fontsize': 20, 'fontweight': 'bold'})

plt.show()
```

## Observation
### Category C1
#### Strong negative correlation of Upper Primary education dropout rate with Per Capita GSDP
#### Strong positive correlation of Senior Secondary education dropout rate with Per Capita GSDP

### Category C2
#### High negative correlation of Primary, Upper Primary, Senior Secondary education dropout rate with Per Capita GSDP

### Category C3
#### High negative correlation of Secondary education dropout rate with Per Capita GSDP

### Category C4
#### Positive correlation of education dropout rate with Per Capita GSDP

## Inference - 
### - High GDP states have strong correlation with Per Capita GSDP
### - Lowest GDP states have positive correlation with Per Capita GSDP
### - For states to move from low GDP to high GDP, they should reduce education dropout rates
